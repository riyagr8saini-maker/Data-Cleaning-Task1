```python
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv("Combined_dataset-checkpoint.csv")
df.head()
```

```python
#Exploring data
print(df.shape)        
print(df.columns)      
print(df.dtypes)       
df.tail()             
```

```python
#Handle Missing Values
# Convert final_price column to numeric (turns non-numeric into NaN)
df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce')

# Show missing values count
print("Missing values before cleaning:")
print(df.isnull().sum())

# Fill missing final_price with mean
df['final_price'] = df['final_price'].fillna(df['final_price'].mean())

# Drop any remaining rows with missing values (optional)
df = df.dropna()

# Confirm changes
print("Missing values after cleaning:")
print(df.isnull().sum())
print("Data type of final_price:", df['final_price'].dtype)

# Preview cleaned data
df.head()

```

```python
#Doing Basic Operations
# Select specific columns
df[['title', 'final_price']].head()

# Filter rows 
df[df['final_price'] > 1000].head()
```

```python
# Remove Duplicates
df = df.drop_duplicates()
```

```python
# Creating derived column
if 'quantity' in df.columns:
    df['total_amount'] = df['final_price'] * df['quantity']
else:
    df['total_amount'] = df['final_price']
```

```python
# Saving cleaned dataset
df.to_csv(r"C:\Users\lenovo\Celebal-Internship task\Task 1\shopping_cleaned.csv", index=False)

```
